# Alpha Shape

```python
from shapely.ops import cascaded_union, polygonize # to install shapely https://conda-forge.github.io/#page-top
from scipy.spatial import Delaunay
import shapely.geometry as geometry
import numpy as np
import math
from matplotlib.collections import LineCollection
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
from descartes import PolygonPatch # to install shapely https://conda-forge.github.io/#page-top
```

```python
# TUTORIAL
# http://blog.thehumangeo.com/2014/05/12/drawing-boundaries-in-python/

# FUNCTIONS

# what is an alpha shape? https://en.wikipedia.org/wiki/Alpha_shape
def alpha_shape(points, alpha):
    if len(points) < 4:
        # When you have a triangle, there is no sense
        # in computing an alpha shape.
        return geometry.MultiPoint(list(points)).convex_hull
    def add_edge(edges, edge_points, coords, i, j):
        if (i, j) in edges or (j, i) in edges:
                # already added
                return
        edges.add( (i, j) )
        edge_points.append(coords[ [i, j] ])
    coords = points
    tri = Delaunay(coords)
    edges = set()
    edge_points = []
    
    # loop over triangles:
    # ia, ib, ic = indices of corner points of the
    # triangle
    for ia, ib, ic in tri.vertices:
        pa = coords[ia]
        pb = coords[ib]
        pc = coords[ic]
        # Lengths of sides of triangle
        a = math.sqrt((pa[0]-pb[0])**2 + (pa[1]-pb[1])**2)
        b = math.sqrt((pb[0]-pc[0])**2 + (pb[1]-pc[1])**2)
        c = math.sqrt((pc[0]-pa[0])**2 + (pc[1]-pa[1])**2)
        # Semiperimeter of triangle
        s = (a + b + c)/2.0
        # Area of triangle by Heron's formula
        area = math.sqrt(s*(s-a)*(s-b)*(s-c))+0.0005
        circum_r = a*b*c/(4.0*area)
        # Here's the radius filter.
        #print circum_r
        if circum_r < 1.0/alpha:
            add_edge(edges, edge_points, coords, ia, ib)
            add_edge(edges, edge_points, coords, ib, ic)
            add_edge(edges, edge_points, coords, ic, ia)
    m = geometry.MultiLineString(edge_points)
    triangles = list(polygonize(m))
    return cascaded_union(triangles), edge_points

# the best way is to create a dictionary of lists
# where keys will be the labels and values will be lists of the points creating the boundaries
# so whenever we want something we can just call the key of the dictionary
from collections import defaultdict

# lets first create a function that returns the point list for a cluster (it is based on the above workflow)
def concave_cluster_points_list(label,method): # label is an integer of a cluster
    # choose the cluster that you wish
    concave_df = df.loc[df[str(method)] == label]
    # this is a two-dimensional numpy array of the points that belong to a cluster
    concave_points = concave_df.as_matrix(columns=['lat', 'lon'])[:,::-1]
    # use the function alpha_shape that we made to get the polygon
    # if you get error "multi-part geometries..." means that the polygon of the cluster has been divided due to alpha value
    concave_hull, edge_points = alpha_shape(concave_points,alpha=alpha_val) # this is the alpha that you set in the beginning  
    return list(concave_hull.boundary.coords)
```

# Hierarchical clustering

```python
# TUTORIAL
# https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/

# needed imports
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

# set up an array with the first class so we can start adding below
X = np.array([new_df.loc[("c0")]])
# we need to make an array with all the points/clusters
for i in range(1,len(new_df)):
    label_array_points = np.array([new_df.loc[("c%d" % i)]]) # you need to have both [[___]] to make it work
    X = np.concatenate((X, label_array_points),axis=0)

# generate the linkage matrix
Z = linkage(X, 'ward')
print "[idx1, idx2, dist, sample_count]\n"
print Z[:10] # print the first 5 to check

# check the Cophenetic Correlation Coefficient of your clustering with help of the cophenet() function.
# The closer the value is to 1, the better the clustering preserves the original distances
c, coph_dists = cophenet(Z, pdist(X))
print c
```

```python
# plotting a dendrogram

'''
horizontal lines are cluster merges
vertical lines tell you which clusters/labels were part of merge forming that new cluster
heights of the horizontal lines tell you about the distance that needed to be "bridged" to form the new cluster
'''

# calculate full dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
pylab.savefig('C:\Users\Sergios\Desktop\Dendrogram.pdf', bbox_inches='tight')

plt.show()
```

```python
# plotting a distance cut-off

def fancy_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        plt.title('Hierarchical Clustering Dendrogram (truncated)')
        plt.xlabel('sample index or (cluster size)')
        plt.ylabel('distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c=c)
                plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                             textcoords='offset points',
                             va='top', ha='center')
        if max_d:
            plt.axhline(y=max_d, c='k')
    return ddata

fancy_dendrogram(
    Z,
    truncate_mode='lastp', # show only the last p merged clusters
    p=40, # show only the last p merged clusters
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,
    annotate_above=10,  # useful in small plots so annotations don't overlap
    max_d=150, # Selecting a Distance Cut-Off aka Determining the Number of Clusters
)
plt.show()
```

```python
# retrieving the clusters below the cut off

from scipy.cluster.hierarchy import fcluster

hier_clusters = fcluster(Z, max_d, criterion='distance')
print hier_clusters.max(),"clusters below the cut-off distance"
print hier_clusters

# make a dataframe with clusters made below the cut-off and the neighbourhood labels
hier_cluster_output = pd.DataFrame({'labels':new_df.index.tolist() , 'cluster':hier_clusters})
hier_cluster_output
```
