```python
# import the libraries to be used
from sklearn.cluster import DBSCAN
from sklearn import metrics

# https://github.com/scikit-learn-contrib/hdbscan/blob/master/README.rst
import hdbscan

# RECAP
# we have one dataframe called coords
# and we are using either array: coords_ar OR scaled
```

### DBSCAN from geoffboeing.com

```python
# http://geoffboeing.com/2014/08/clustering-to-reduce-spatial-data-set-size/

# epsilon and coordinates get converted to radians, because haversine metric needs radian units
kms_per_radian = 6371.0088
epsilon = 0.05 / kms_per_radian

# choose the data
data = coords_ar

# create a DBSCAN cluster object
db = DBSCAN(eps=epsilon, min_samples=1.5, algorithm='ball_tree', metric='haversine').fit(np.radians(data))

# pass the labels in the original dataframe
dbscan1_labels = db.labels_
coords['labels']=dbscan1_labels

# test the coords dataframe and print the max and min labels
print "COORDS DATAFRAME with LABELS"
print coords.head(), "\n"

# Clusters info
num_clusters = len(set(dbscan1_labels))
clusters = pd.Series([data[dbscan1_labels == n] for n in range(num_clusters)])
print('Number of clusters: {}'.format(num_clusters))

# extract the Silhouette Score for the clusters created above on the data used.
print metrics.silhouette_score(data, dbscan1_labels)

# check how many outliers are there
outliers = len(coords.loc[coords['labels'] == -1])
print "Outliers:", outliers, "out of", len(coords), "data points ie.",outliers*100/len(coords),"%"

coords.plot.scatter(x='lon', y='lat', c='labels',figsize=(30,15),cmap='hsv', alpha=0.5) 
plt.title('OSM amenities plot', fontsize=30)
plt.ylabel('latitude', fontsize=26)
plt.xlabel('longitude', fontsize=26)
plt.show()
```

### HDBSCAN 1

```python
# as seen at http://hdbscan.readthedocs.io/en/latest/comparing_clustering_algorithms.html
import sklearn.cluster as cluster
import time
%matplotlib inline
sns.set_context('poster')
sns.set_color_codes()
plot_kwds = {'alpha' : 0.25, 's' : 80, 'linewidths':0}


def plot_clusters(data, algorithm, args, kwds):
    start_time = time.time()
    labels = algorithm(*args, **kwds).fit_predict(data)
    end_time = time.time()
    palette = sns.color_palette('deep', np.unique(labels).max() + 1)
    colors = [palette[x] if x >= 0 else (0.0, 0.0, 0.0) for x in labels]
    plt.scatter(data.T[1], data.T[0], c=colors, **plot_kwds)
    frame = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)
    plt.title('Clusters found by {}'.format(str(algorithm.__name__)), fontsize=24)
```

```python
# make a new dataframe with max rows as number of clusters-1(without outliers) and columns the unique values of "amenities"
num_clusters = len(set(hdbscan2_labels))
print "The biggest label is", num_clusters, "therefore we need", num_clusters-1, "rows, taking out the cluster of the outliers"
columns = []
for i in range(0,num_clusters):
    a = df["amenity"][i]
    if a not in columns:
        columns.append(a)
        
new_df = pd.DataFrame(index=range(0,num_clusters),columns=columns)
new_df
```

```python
def count_amenities(amenity):
    for i in range(0,num_clusters):
        def_df = df.loc[df['labels'] == i]
        new_df[str(amenity)][i] = len(def_df.loc[def_df['amenity'] == str(amenity)])
    #return new_df

for c in columns:
    count_amenities(str(c))
```
